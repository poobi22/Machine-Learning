```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
a=pd.read_csv("homeprices.csv")
a
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('prices')
plt.scatter(a.area,a.price)
```

```python
r=linear_model.LinearRegression()
r.fit(a[['area']],a.price)
```

```python
r.predict([[3300]])
```

```python
r.predict([[4000]])
```

```python
b=pd.read_csv("area.csv")
b
```

```python
price=r.predict(b[['area']])
price
```

```python
b['prices']=price
b
```
