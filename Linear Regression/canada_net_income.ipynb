```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
c=pd.read_csv("canada_per_capita_income.csv")
c
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(c.year,c['per capita income (US$)'])
```

```python
r=linear_model.LinearRegression()
r.fit(c[['year']],c['per capita income (US$)'])
```

```python
r.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(c.year,c['per capita income (US$)'])
plt.plot(c.year,r.predict(c[['year']]))
```
